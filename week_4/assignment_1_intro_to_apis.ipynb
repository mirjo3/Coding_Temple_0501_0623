### Problem 3:

Create a function that does the following:
* Input: pokemon_name(user_defined) -> str
* Output: list object containing a dictionary of values from the response object.
* Please include the type, the most current raritey level, at least one held item, and all stats

```python
import requests
from bs4 import BeautifulSoup

def get_pokemon_info(pokemon_name):
    page_poke = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}")
    poke_json = page_poke.json()
    
    pokemon_info = {
        'name': poke_json['name'],
        'type': poke_json['types'][0]['type']['name'],
        'rarity': poke_json['held_items'][0]['version_details'][0]['rarity'],
        'held_item': poke_json['held_items'][0]['item']['name'],
        'stats': poke_json['stats'][0]['base_stat']
    }
    
    return pokemon_info

# Example usage
pokemon_name = input("Enter a Pokémon name: ")
pokemon_info = get_pokemon_info(pokemon_name)
print('Name:', pokemon_info['name'].title())
print('Type:', pokemon_info['type'])
print('Rarity:', pokemon_info['rarity'])
print('Held Items:', pokemon_info['held_item'])
print('Stats:', pokemon_info['stats'])
```
