- Create a user-defined function that allows a user to input a category and returns a visualization of your choice.

```python
def cate_id(data):
    fig, ax = plt.subplots()
    
    # Customize the plot
    ax.set_title('Histogram')
    ax.set_xlabel('Category IDs')
    ax.set_ylabel('Count')
    
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    # Create the histogram using the provided data
    ax.hist(data, bins=50)
    
    plt.show()

data = df['category_parent_id'].value_counts()
cate_id(data)
```

### Task 2: Load the data
Write a function which does the following:
- Input: 
    - Function should take a filepath (str) object
- Creates a dataframe using pandas
- Lower-case all column headers
- Replace the space with an underscore (_) in all column headers
- Output:
    - Cleaned dataframe object

```python
def clean_dataframe(filepath):
    df = pd.read_csv(filepath, encoding='utf-8')
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    
    return df

filepath = r'C:\Users\jmira\OneDrive\Coding Temple Cohort 0501 0623\week_4\Food_Costs_2016.csv'
cleaned_df = clean_dataframe(filepath)

pd.set_option('display.max_rows', None)
cleaned_df['cumulative_cost'] = pd.to_numeric(cleaned_df['cumulative_cost'], errors='coerce')
total_df = cleaned_df['cumulative_cost'].sum()
```

```python
def clean_dataframe(filepath):
    df = pd.read_csv(filepath, encoding='utf-8')
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    
    return df

filepath = r'C:\Users\jmira\OneDrive\Coding Temple Cohort 0501 0623\week_4\Food_Costs_2015.csv'
df_total_cost_2015 = clean_dataframe(filepath)

# .merge() 	state_agency_or_indian_tribal_organization to get participants and cost
merged_df_total = pd.merge(df_total_cost_2016, df_total_cost_2015, on='state_agency_or_indian_tribal_organization')

selected_df_total = merged_df_total[['state_agency_or_indian_tribal_organization', 'cumulative_cost_x', 'cumulative_cost_y']]

# Renaming new columns to compare fiscal years
selected_df_total = selected_df_total.rename(columns={
    'cumulative_cost_x': 'cumulative_cost_2016', 
    'cumulative_cost_y': 'cumulative_cost_2015'
})

sorted_df_total = selected_df_total.sort_values(by='state_agency_or_indian_tribal_organization', ascending=True).reset_index(drop=True)
sorted_df_total
```
