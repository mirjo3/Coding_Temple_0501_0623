### #5 Which neighorhood group has the highest average price?

```python
# I want to target neighbourhood_group and prices
# Find a way to get the mean() of eah neighbourhood_group
#bnb_price = bnb.sort_values("price", ascending = False)
#display(bnb_price)
# Due to the complexity of the data, there are many outliers when we sort by price alone.
# Example, 'host_id' 20582832, 'price' 10,000, 'minimum_nights' 100, 'availability_365' 0
# Based on this data, you would need 1,000,000 to spend at her place for 100 nights and not available anytime of the year
# We will need a conditional mask to filter out long term rentals vs short term rentals
# For short term rentals we will look for < 5 days.
# For long term rentals we will look at > 5 days

# Short term Rental availability less than 5 days, under 9000
st_conditional_mask = bnb['minimum_nights'] <= 5
st_conditional_price = bnb['price'] <= 9000
stc_bnb = bnb[st_conditional_mask & st_conditional_price]
stp_bnb = stc_bnb.sort_values("price", ascending=False)
print("\nTop 5 short term (under 5 nights) stay rentals")

# We use .head() to only isolate the top 5
display(stp_bnb.head(5)[['neighbourhood_group', 'price']])



# Long term Rental availability for more than 5 days, under 9000
lt_conditional_mask = bnb['minimum_nights'] > 5
lt_conditional_price = bnb['price'] <= 9000
ltc_bnb = bnb[lt_conditional_mask & lt_conditional_price]
ltp_bnb = ltc_bnb.sort_values("price", ascending=False)
print("\nTop 5 long term (over 5 nights) stay rentals")

# We use .head() to only isolate the top 5
display(ltp_bnb.head(5)[['neighbourhood_group', 'price']])
```

### #6. Which neighborhood group has the highest total price?

```python
# Using the data from question 5, we will take the top 10 of each
# short term and long term facility and taking the average

# Calculate and display average price by neighbourhood_group
stp_avg = stp_bnb.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("\nAverage price of short term rentals by neighbourhood_group:\n")
print(stp_avg)

# Calculate and display average price by neighbourhood_group
ltp_avg = ltp_bnb.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("\nAverage price of long term rentals by neighbourhood_group:\n")
print(ltp_avg)

# Caveats to consider: 
# To gain better insight on the data we should be filtering by 'room_type' for a more accurate comparison
```
